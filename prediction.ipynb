## Step2: 可视化-职位和相关职能的关系
### 1）选取相关特征
### 2）整理数据

```python
position_names = content['positionName'].to_list()
skill_labels = content['skillLables'].to_list()
skill_labels
```

```python
from collections import defaultdict
```

```python
position_skill_graph = defaultdict(list)
```

```python
for p, s in zip(position_names, skill_labels):
    # 将s从str转变为list，便于相加
    position_skill_graph[p] += eval(s)
print(len(position_skill_graph))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph(position_skill_graph)
```

```python
#设置中文字体为黑体
plt.rcParams['font.sans-serif'] = ['SimHei']
#用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
#加载字体
from matplotlib.font_manager import _rebuild
_rebuild() #reload一下
```

```python
#随机取5个职位做可视化，截取出5个职位和其相关的技能节点
import random
sample_positions = random.sample(position_names, k=5)
print(sample_positions)
```

```python
sample_nodes_connections = sample_positions
```

```python
for p, s in position_skill_graph.items():
    if p in sample_positions:
        sample_nodes_connections += s
```

```python
sample_graph = G.subgraph(sample_nodes_connections)
```

```python
#设置nodes间距离 --> k
pos = nx.spring_layout(sample_graph, k=1)
```

```python
nx.draw(sample_graph, pos=pos, with_labels=True, hold=False)
```

```python
#使用pagerank算法，对核心能力和核心职位排序
pr = nx.pagerank(G, alpha=0.9)
ranked_position_and_ability = sorted([(name, value) for name, value in pr.items()], key=lambda x:x[1], reverse=True)
print(ranked_position_and_ability[:5])
```

## Step3: 数据清洗

```python
X_content = content.drop(['salary'], axis=1)
```

```python
Y_content = content['salary'].to_list()
```

```python
X_content['merged'] = X_content.apply(lambda x: ''.join(str(x)), axis=1)
```

```python
X_string = X_content['merged'].to_list()
print(X_string[0])
```

```python
#数据预处理
import jieba
import re
```

```python
#过滤特殊符号
def token(string):
    return re.findall('\w+', string)
```

```python
def get_one_row_string(x_row_string):
    job_string = ''
    for i, element in enumerate(x_row_string.split('\n')):
        if len(element.split()) == 2:
            #舍弃id字段
            if i == 0: continue
            _, value = element.split()
            job_string += value
    return job_string
```

```python
job_string = get_one_row_string(X_string[0])
job_string
```

```python
cutted_X = []
```

```python
for i, row in enumerate(X_string):
    job_string = get_one_row_string(row)
    if i % 1000 == 0:
        print(i)
    #对句子进行分词, 并以空格隔开连接
    cutted_X.append(' '.join(list(jieba.cut(''.join(token(job_string))))))
print(cutted_X[0])
```

```python
import numpy as np
```

```python
Y_numerical = [np.mean(list(map(float, re.findall('\d+', s)))) for s in Y_content]
Y_numerical
```

```python
print("length of X: {}, length of Y: {}".format(len(cutted_X), len(Y_numerical)))
```

## Step4: 提取文本特征TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(cutted_X)
```

```python
Y = Y_numerical
```

## Step5: 回归分析，使用KNN回归训练模型

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
```

```python
knn_model = KNeighborsRegressor(n_neighbors=2)
```

```python
knn_model.fit(X, Y)
```

```python
def predict_by_description(description, model):
    test_words = list(jieba.cut(description))
    test_vec = vectorizer.transform(test_words)
    pred = model.predict(test_vec)
    return pred[0]
```
